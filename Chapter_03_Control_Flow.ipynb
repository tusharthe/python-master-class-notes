# Control Flow




### If Statement (Slide 26)


```python
# Basic if statement
age = 18

if age >= 18:
    print("You are an adult")
    print("You can vote")

print("Program continues...")
```

> **Note:** Code block must be indented


### If-Else (Slide 27)


```python
# If-else statement
score = 75

if score >= 60:
    print("Pass")
    grade = "Passed"
else:
    print("Fail")
    grade = "Failed"

print(f"Result: {grade}")
```

### If-Elif-Else (Slide 28)


```python
# Multiple conditions
score = 85

if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
elif score >= 70:
    grade = "C"
elif score >= 60:
    grade = "D"
else:
    grade = "F"

print(f"Your grade: {grade}")
```

> **Note:** Checks conditions top to bottom

### Logical Operators (Slide 30)


```python
# and, or, not
age = 25
has_license = True

# and - both must be True
if age >= 18 and has_license:
    print("Can drive")

# or - at least one True
if age < 18 or not has_license:
    print("Cannot drive")

# not - reverses boolean
if not False:
    print("This runs")
```

### Nested If Statements (Slide 31)


```python
# If inside if
age = 20
has_license = True
has_insurance = False

if age >= 18:
    if has_license:
        if has_insurance:
            print("Can drive legally")
        else:
            print("Need insurance")
    else:
        print("Need license")
else:
    print("Too young")
```

### Ternary Operator (Slide 32)


```python
# Conditional expression (one-liner)
age = 20
status = "Adult" if age >= 18 else "Minor"
print(status)  # Adult

# Another example
x = 10
result = "Even" if x % 2 == 0 else "Odd"
print(result)  # Even

# Can be nested (but avoid complexity)
num = 15
category = "positive" if num > 0 else "negative" if num < 0 else "zero"
print(category)  # positive
```

> **Note:** Great for simple conditions


### While Loop (Slide 33)


```python
# Repeat while condition is True
count = 0
while count < 5:
    print(count)
    count += 1  # Increment

# Output: 0, 1, 2, 3, 4

# Without increment = infinite loop!
# while True:
#     print("Forever!")  # DON'T RUN THIS
```

> **Note:** Always ensure loop will end


### While with Break (Slide 34)


```python
# break exits the loop immediately
count = 0
while True:  # Infinite loop
    print(count)
    count += 1
    if count >= 5:
        break  # Exit loop

print("Loop ended")

# Useful for user input
while True:
    # answer = input("Continue? (y/n): ")  # (Interactive input disabled)
    answer = 'n'  # Mocked input
    if answer == 'n':
        break
```

### While with Continue (Slide 35)


```python
# continue skips to next iteration
count = 0
while count < 10:
    count += 1
    if count % 2 == 0:
        continue  # Skip even numbers
    print(count)

# Output: 1, 3, 5, 7, 9

# Only odd numbers printed
```

> **Note:** continue skips rest of current iteration


### For Loop - Lists (Slide 36)


```python
# Iterate over a list
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)

# Output:
# apple
# banana  
# cherry

# With index
for i in range(len(fruits)):
    print(f"{i}: {fruits[i]}")
```

### For Loop - Range (Slide 37)


```python
# range(stop)
for i in range(5):
    print(i)  # 0, 1, 2, 3, 4

# range(start, stop)
for i in range(2, 6):
    print(i)  # 2, 3, 4, 5

# range(start, stop, step)
for i in range(0, 10, 2):
    print(i)  # 0, 2, 4, 6, 8

# Countdown
for i in range(5, 0, -1):
    print(i)  # 5, 4, 3, 2, 1
```

> **Note:** range is very efficient


### For Loop - Enumerate (Slide 38)


```python
# Get index and value together
fruits = ["apple", "banana", "cherry"]

for index, fruit in enumerate(fruits):
    print(f"{index}: {fruit}")

# Output:
# 0: apple
# 1: banana
# 2: cherry

# Start index from 1
for index, fruit in enumerate(fruits, start=1):
    print(f"{index}. {fruit}")
```

> **Note:** Better than range(len(...))


### Nested Loops (Slide 39)


```python
# Loop inside a loop
for i in range(3):
    for j in range(2):
        print(f"i={i}, j={j}")

# Multiplication table
for i in range(1, 4):
    for j in range(1, 4):
        print(f"{i} x {j} = {i*j}", end="  ")
    print()  # New line after each row
```

### Loop Else Clause (Slide 40)


```python
# else runs if loop completes without break
for i in range(5):
    if i == 10:  # Never true
        break
else:
    print("Loop completed normally")

# With break
for i in range(5):
    if i == 3:
        break
else:
    print("This won't print")

print("After loop")
```

> **Note:** Useful for search operations
