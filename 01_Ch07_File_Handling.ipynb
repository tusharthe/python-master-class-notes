# File Handling




### Opening Files (Slide 105)


```python
# Open file for reading
file = open("data.txt", "r")
content = file.read()
print(content)
file.close()  # Always close!

# Better: use 'with' (auto-closes)
with open("data.txt", "r") as file:
    content = file.read()
    print(content)
# File automatically closed here

# File modes:
# "r" - read (default)
# "w" - write (overwrites)
# "a" - append
# "r+" - read and write
```

> **Note:** Always use 'with' to auto-close files


### Reading Files - read() (Slide 106)


```python
# Read entire file
with open("data.txt", "r") as file:
    content = file.read()
    print(content)

# Read specific number of bytes
with open("data.txt", "r") as file:
    first_10 = file.read(10)
    print(first_10)

# Check if file is closed
print(file.closed)  # True

# Read from current position
with open("data.txt", "r") as file:
    print(file.read(5))   # First 5 chars
    print(file.read(5))   # Next 5 chars
```

> **Note:** read() loads entire file into memory


### Reading Files - readline() (Slide 107)


```python
# Read one line at a time
with open("data.txt", "r") as file:
    line1 = file.readline()
    line2 = file.readline()
    print(line1)
    print(line2)

# Read all lines into list
with open("data.txt", "r") as file:
    all_lines = file.readlines()
    print(all_lines)  # ['line1\n', 'line2\n', ...]

# Memory efficient: iterate
with open("data.txt", "r") as file:
    for line in file:
        print(line.strip())  # Remove \n
```

> **Note:** Iterating is most memory-efficient


### Writing Files (Slide 108)


```python
# Write mode - overwrites file
with open("output.txt", "w") as file:
    file.write("Hello, World!\n")
    file.write("Second line\n")

# Write multiple lines
lines = ["Line 1\n", "Line 2\n", "Line 3\n"]
with open("output.txt", "w") as file:
    file.writelines(lines)

# Write with print
with open("output.txt", "w") as file:
    print("Hello", file=file)
    print("World", file=file)

# Note: write() doesn't add \n automatically!
```

> **Note:** w mode creates file if it doesn't exist


### Appending to Files (Slide 109)


```python
# Append mode - adds to end
with open("log.txt", "a") as file:
    file.write("New log entry\n")

# Create if doesn't exist
with open("new_file.txt", "a") as file:
    file.write("First line\n")

# Append multiple lines
with open("log.txt", "a") as file:
    entries = ["Entry 1\n", "Entry 2\n"]
    file.writelines(entries)

# Common pattern: logging
import datetime
with open("app.log", "a") as log:
    timestamp = datetime.datetime.now()
    log.write(f"[{timestamp}] App started\n")
```

> **Note:** a mode never overwrites existing content


### File Exists Check (Slide 110)


```python
import os

# Check if file exists
if os.path.exists("data.txt"):
    print("File exists!")
else:
    print("File not found")

# Check if it's a file (not directory)
if os.path.isfile("data.txt"):
    print("It's a file")

# Check if directory
if os.path.isdir("my_folder"):
    print("It's a directory")

# Get file size
size = os.path.getsize("data.txt")
print(f"Size: {size} bytes")

# Safe file reading
filename = "data.txt"
if os.path.exists(filename):
    with open(filename, "r") as f:
        print(f.read())
```

### pathlib - Modern Way (Slide 112)


```python
from pathlib import Path

# Create path object
file_path = Path("data") / "files" / "data.txt"
print(file_path)

# Check if exists
if file_path.exists():
    print("File exists")

# Read file
content = file_path.read_text()
print(content)

# Write file
file_path.write_text("Hello, World!")

# Get parts
print(file_path.name)      # data.txt
print(file_path.stem)      # data
print(file_path.suffix)    # .txt
print(file_path.parent)    # data/files

# Create directory
Path("new_folder").mkdir(exist_ok=True)
```

> **Note:** pathlib is more Pythonic than os.path


### Reading CSV Files (Slide 113)


```python
import csv

# Read CSV
with open("data.csv", "r") as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        print(row)  # List of values

# Read with headers
with open("data.csv", "r") as file:
    csv_reader = csv.DictReader(file)
    for row in csv_reader:
        print(row["name"], row["age"])

# Write CSV
data = [
    ["name", "age", "city"],
    ["Alice", 25, "NYC"],
    ["Bob", 30, "LA"]
]
with open("output.csv", "w", newline="") as file:
    csv_writer = csv.writer(file)
    csv_writer.writerows(data)
```

> **Note:** CSV module handles commas in data


### Writing CSV with DictWriter (Slide 114)


```python
import csv

# Write dict data to CSV
data = [
    {"name": "Alice", "age": 25, "city": "NYC"},
    {"name": "Bob", "age": 30, "city": "LA"},
    {"name": "Charlie", "age": 35, "city": "SF"}
]

with open("people.csv", "w", newline="") as file:
    fieldnames = ["name", "age", "city"]
    writer = csv.DictWriter(file, fieldnames=fieldnames)

    writer.writeheader()  # Write column names
    writer.writerows(data)  # Write all rows

# Result:
# name,age,city
# Alice,25,NYC
# Bob,30,LA
# Charlie,35,SF
```

> **Note:** DictWriter is cleaner for structured data


### Reading JSON Files (Slide 115)


```python
import json

# Read JSON file
with open("data.json", "r") as file:
    data = json.load(file)
    print(data)  # Python dict

# Example JSON file content:
# {"name": "Alice", "age": 25, "hobbies": ["reading", "coding"]}

# Access data
print(data["name"])
print(data["hobbies"][0])

# Handle errors
try:
    with open("data.json", "r") as file:
        data = json.load(file)
except json.JSONDecodeError:
    print("Invalid JSON!")
except FileNotFoundError:
    print("File not found!")
```

> **Note:** JSON maps directly to Python dicts/lists


### Writing JSON Files (Slide 116)


```python
import json

# Python data
person = {
    "name": "Alice",
    "age": 25,
    "city": "NYC",
    "hobbies": ["reading", "coding"]
}

# Write to file
with open("person.json", "w") as file:
    json.dump(person, file, indent=2)

# Result (pretty-printed):
# {
#   "name": "Alice",
#   "age": 25,
#   "city": "NYC",
#   "hobbies": ["reading", "coding"]
# }

# Without indent (compact)
with open("compact.json", "w") as file:
    json.dump(person, file)
```

### File Exceptions (Slide 117)


```python
# Handle file errors
try:
    with open("missing.txt", "r") as file:
        content = file.read()
except FileNotFoundError:
    print("File doesn't exist!")
except PermissionError:
    print("No permission to read file")
except Exception as e:
    print(f"Error: {e}")

# Safe file write
filename = "output.txt"
try:
    with open(filename, "w") as file:
        file.write("Hello")
    print("File written successfully")
except IOError as e:
    print(f"Write failed: {e}")

# Always use try-except for file operations!
```

> **Note:** File operations can fail in many ways


### Binary Files (Slide 118)


```python
def process(x): pass
# Read binary file (images, etc.)
with open("image.jpg", "rb") as file:
    binary_data = file.read()
    print(len(binary_data))  # Size in bytes

# Write binary
with open("copy.jpg", "wb") as file:
    file.write(binary_data)

# Copy file
def copy_file(source, dest):
    with open(source, "rb") as src:
        with open(dest, "wb") as dst:
            dst.write(src.read())

copy_file("original.jpg", "backup.jpg")

# Read in chunks (large files)
with open("large.bin", "rb") as file:
    while chunk := file.read(1024):  # 1KB chunks
        process(chunk)
```

> **Note:** Use 'b' mode for non-text files


### File Context Manager (Slide 119)


```python
# 'with' statement ensures file closes
with open("data.txt", "r") as file:
    content = file.read()
# File auto-closed even if error occurs!

# Multiple files
with open("input.txt", "r") as infile,      open("output.txt", "w") as outfile:
    content = infile.read()
    outfile.write(content.upper())

# Without 'with' (not recommended)
file = open("data.txt", "r")
try:
    content = file.read()
finally:
    file.close()  # Must close manually

# 'with' is always better!
```

> **Note:** with automatically calls file.close()


### File Modes Summary (Slide 120)


<p><strong>File Opening Modes:</strong></p>
<ul>
<li><strong>r</strong> - Read (default), error if file doesn't exist</li>
<li><strong>w</strong> - Write, creates file, overwrites existing</li>
<li><strong>a</strong> - Append, creates file, adds to end</li>
<li><strong>x</strong> - Exclusive create, error if file exists</li>
<li><strong>r+</strong> - Read and write</li>
<li><strong>w+</strong> - Write and read (overwrites)</li>
<li><strong>a+</strong> - Append and read</li>
</ul>
<p><strong>Binary mode:</strong> Add 'b' (rb, wb, ab)</p>
<p><strong>Text mode:</strong> Add 't' (rt, wt) - default</p>

### Temporary Files (Slide 122)


```python
import tempfile

# Create temporary file
with tempfile.TemporaryFile(mode="w+") as temp:
    temp.write("Temporary data")
    temp.seek(0)  # Go to beginning
    print(temp.read())
# File auto-deleted when closed

# Named temporary file
with tempfile.NamedTemporaryFile(mode="w", delete=False) as temp:
    temp.write("Data")
    temp_name = temp.name
    print(f"Temp file: {temp_name}")

# Temporary directory
with tempfile.TemporaryDirectory() as tmpdir:
    print(f"Temp dir: {tmpdir}")
    # Use tmpdir...
# Directory auto-deleted
```

> **Note:** Useful for testing and caching


### File Iteration Best Practices (Slide 123)


```python
# Process large files line by line
def count_lines(filename):
    count = 0
    with open(filename, "r") as file:
        for line in file:  # Memory efficient!
            count += 1
    return count

# Filter lines
def filter_comments(input_file, output_file):
    with open(input_file, "r") as infile,          open(output_file, "w") as outfile:
        for line in infile:
            if not line.strip().startswith("#"):
                outfile.write(line)

# Process in chunks
def process_large_file(filename, chunk_size=1024):
    with open(filename, "rb") as file:
        while chunk := file.read(chunk_size):
            # Process chunk
            yield chunk
```

> **Note:** Never read entire large file into memory
